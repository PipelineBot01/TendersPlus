```python
import pandas as pd, numpy as np
```

```python
relation_df = pd.read_csv('data/tag_category_map.csv')
staff_df = pd.read_csv('data/clean_staff_tags.csv')
```

```python
merge_df = staff_df.merge(relation_df)
```

```python
merge_df.groupby('value')['Staff ID'].count().reset_index().sort_values('Staff ID', ascending=False)
```

```python
cate_df = merge_df.groupby(['Staff ID','value'])['Weight'].sum().reset_index().sort_values('Weight', ascending=False)
cleaned_cate_df = cate_df[~cate_df['value'].isin(['DIVISION 24 OTHERS(IRRELEVANT)', 'DIVISION 23 OTHERS(RELEVANT)'])]
cleaned_cate_df = cleaned_cate_df.groupby(['Staff ID']).apply(lambda x:x.sort_values(by=['Weight'],ascending=False)[:1])
```

```python
result = cleaned_cate_df.groupby('value')['Staff ID'].count().reset_index().sort_values('Staff ID', ascending=False).rename(columns={'value':'category'})
display(result)
```

```python
def check(df):
    df['Weight'] = df['Weight'] / (df['Weight'].sum())
    print(df)
    return df
```
