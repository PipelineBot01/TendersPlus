```python
import pandas as pd, numpy as np
```

```python
rel_df = pd.read_csv('Data/relevant.csv')
oppo_df = pd.read_csv('Data/opportunities.csv')
```

```python
merge_df = rel_df.append(oppo_df)
```

```python
oppo_df = oppo_df.rename(columns={'Primary Category': 'Category', 
                                  'Secondary Category': 'Category_sub', 
                                  'Estimated Grant Value (AUD)': 'Value',
                                  'Close Date & Time': 'Close Date'})
```

```python
oppo_df.columns
```

```python
tmp_df = oppo_df[['_id', 'Agency', 'Title', 'Publish Date', 'Close Date', 'Category', 'Category_sub', 'Description', 'Eligibility', 'Value', 'Location']]
```

```python
tmp_df
```

```python
tmp_df['Publish Date'] = pd.to_datetime(tmp_df['Publish Date'])
tmp_df['Close Date'] = pd.to_datetime(tmp_df['Close Date'], errors='coerce').dt.date
```

```python
import re
```

```python
def extract_money(item):
    if type(item) == type(1.01): return item
    re_rule = 'From\$(\d+)to\$(\d+)'
    item = item.replace(',', '').replace('.', '')
    result = re.findall(re_rule,item)
    return '-'.join(str(int(i)) for i in result[0])
```

```python
tmp_df['Value'] = tmp_df['Value'].map(lambda x:extract_money(x))
```

```python
merge_df = tmp_df.append(rel_df).drop('ATM ID', axis=1)
```

```python
def remove(x):
    if '-' in x: return x.split('-')[-1][1:]
    return x
```

```python
merge_df['Category'] = merge_df['Category'].map(remove)
```

```python
merge_df.to_csv('tenders_info.csv', index=0, encoding='utf-8_sig')
```
